```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from skimage import io, feature, filters,util
from skimage.color import rgb2gray
from ipywidgets import interact


def normalize(arr):
    arr_min = arr.min()
    arr_max = arr.max()
    return (arr - arr_min) / (arr_max - arr_min)
```

```python
#red = img_norm[:,:,0]
#red_denoised = filters.median(img_norm[:,:,0],selem=np.ones((5,5)))
img_denoised = filters.median(img,selem=np.ones((5,5)))
f,(ax0,ax1) = plt.subplots(1,2,figsize=(10,5))
ax0.imshow(img,cmap='gray')
ax1.imshow(img_denoised,cmap='gray')
```

```python
#busco los bordes de los defectos
edges = feature.canny(img_denoised,sigma=1) #play with sigma
f,(ax0,ax1) = plt.subplots(1,2,figsize=(10,5))
ax0.imshow(img_denoised,cmap='gray')
ax1.imshow(edges,cmap='gray')
```

```python
#creamos fuentes en los defectos
#el color en el mapa nos dice qu tan lejos est cada punto de los bordes de los defectos

from scipy.ndimage import distance_transform_edt
dt = distance_transform_edt(~edges)
f,(ax0,ax1) = plt.subplots(1,2,figsize=(10,5))
ax0.imshow(img_denoised,cmap='gray')
ax1.imshow(dt,cmap='gray')
```

```python
peak_idx =feature.peak_local_max(dt,indices=True,min_distance=2)
plt.plot(peak_idx[:,1],peak_idx[:,0],'r.')
io.imshow(dt,cmap='gray');
```

```python
#buscamos las ubicaciones de los picos de intensidad de las fuentes creadas (defectos)
local_max = feature.peak_local_max(dt,indices=False,min_distance=2)
```

```python
from skimage import measure
markers = measure.label(local_max)
```

```python
from skimage import morphology, segmentation
labels = morphology.watershed(-dt,markers)
io.imshow(segmentation.mark_boundaries(img_denoised,labels));
```

```python
from skimage import color
io.imshow(color.label2rgb(labels,image=img_denoised));
```

```python
io.imshow(color.label2rgb(labels,image=img_denoised,kind='avg'),cmap='gray');
```

```python
regions = measure.regionprops(labels,intensity_image=img_denoised)
print(len(regions))
for prop in regions:
    print('Label:{} diameter:{} microns'.format(prop.label,prop.equivalent_diameter*pixels_to_microns))
```

```python
region_means = [r.mean_intensity for r in regions]
plt.hist(region_means,bins=50);
```

```python
#del histograma de arriba se diferencia el background, los defectos pequeos y el 3er grupo de los defectos grandes
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
region_means = np.array(region_means).reshape(-1,1)
model.fit(region_means)
print(model.cluster_centers_)
```

```python
bg_fg_labels = model.predict(region_means)
bg_fg_labels
```

```python
classified_labels = labels.copy()
for bg_fg, region in zip(bg_fg_labels,regions):
    classified_labels[tuple(region.coords.T)] = bg_fg
```

```python
io.imshow(classified_labels,cmap='gray')
regs = measure.regionprops(classified_labels,intensity_image=img_denoised)
```
