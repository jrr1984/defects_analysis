```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from skimage import io, feature, filters
from ipywidgets import interact


def normalize(arr):
    arr_min = arr.min()
    arr_max = arr.max()
    return (arr - arr_min) / (arr_max - arr_min)
```

```python
pixels_to_microns = 0.586 #1pixel=0.586 microns
```

```python
img = io.imread('C:/Users/juanr/Documents/mediciones microscopio metalografico/amarillo_11.tif')
```

```python
img_norm = normalize(img)
io.imshow(img_norm[:,:,0],cmap='gray');
```

```python
img_norm.shape
```

```python
plt.hist(img_norm[:,:,0].flatten(),bins=100,log=True)
```

```python
red = img_norm[:,:,0]
red_denoised = filters.median(img_norm[:,:,0],selem=np.ones((5,5)))
f,(ax0,ax1) = plt.subplots(1,2,figsize=(10,5))
ax0.imshow(red)
ax1.imshow(red_denoised)
```

```python
#busco los bordes de los defectos
edges = feature.canny(red,sigma=3.5)
io.imshow(edges)
```

```python
#creamos fuentes en los defectos
#el mapa nos dice qué tan lejos está cada punto de los bordes de los defectos

from scipy.ndimage import distance_transform_edt
dt = distance_transform_edt(~edges)
io.imshow(dt);
```

```python
#buscamos las ubicaciones de los picos de intensidad de las fuentes creadas (defectos)
local_max = feature.peak_local_max(dt,indices=False,min_distance=2)
io.imshow(local_max,cmap='gray');
```

```python
peak_idx =feature.peak_local_max(dt,indices=True,min_distance=2)
plt.plot(peak_idx[:,1],peak_idx[:,0],'r.')
io.imshow(dt);
```

```python
from skimage import measure
markers = measure.label(local_max)
```

```python
from skimage import morphology, segmentation
labels = morphology.watershed(-dt,markers)
io.imshow(segmentation.mark_boundaries(red,labels));
```

```python
from skimage import color
io.imshow(color.label2rgb(labels,image=red));
```

```python
io.imshow(color.label2rgb(labels,image=red,kind='avg'),cmap='gray');
```

```python
regions = measure.regionprops(labels,intensity_image=red)
len(regions)
for prop in regions:
    print('Label:{} diameter:{} microns'.format(prop.label,prop.equivalent_diameter*pixels_to_microns))
```

```python
region_means = [r.mean_intensity for r in regions]
plt.hist(region_means,bins=50)
```

```python
#del histograma de arriba se diferencia el background, los defectos pequeños y el 3er grupo de los defectos grandes
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
region_means = np.array(region_means).reshape(-1,1)
model.fit(region_means)
print(model.cluster_centers_)
```

```python
bg_fg_labels = model.predict(region_means)
bg_fg_labels
```

```python
classified_labels = labels.copy()
for bg_fg, region in zip(bg_fg_labels,regions):
    classified_labels[tuple(region.coords.T)] = bg_fg
```

```python
io.imshow(classified_labels)
```

```python
io.imshow(color.label2rgb(classified_labels,image=red),cmap='gray')
```
