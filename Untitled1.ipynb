```python
from skimage import data,feature,io,filters
from matplotlib import pyplot as plt
import math
import numpy as np

img = io.imread('C:/Users/juanr/Documents/mediciones microscopio metalografico/amarillo_11.tif')
img = img[:,:,0]

def normalize(arr):
    arr_min = arr.min()
    arr_max = arr.max()
    return (arr - arr_min) / (arr_max - arr_min)
img = normalize(img)
```

```python
plt.hist(img.flatten(),bins=100,log=True)
```

```python
img_denoised = filters.median(img,selem=np.ones((5,5)))
edges = feature.canny(img_denoised,sigma=1) #play with sigma
io.imshow(edges)
```

```python
#blobs = feature.blob_doh(img,max_sigma=30,threshold=.00001)
blobs = feature.blob_dog(img, max_sigma=1, threshold=0.8)
blobs[:, 2] = blobs[:, 2] * np.sqrt(2)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
plt.imshow(img,cmap='gray')


for b in blobs:
    y, x, r = b[0],b[1],b[2]
    c = plt.Circle((y,x),r,color='#ff0000',lw = 2,fill = False)
    ax.add_patch(c)

plt.show()
```
